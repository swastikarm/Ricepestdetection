```
%%writefile app.py

import streamlit as st
import numpy as np
import cv2
from PIL import Image
import tempfile
from roboflow import Roboflow

# Initialize Roboflow with API key
rf = Roboflow(api_key="nfnOVPYna2umvXZnJp2I")

# Load project and version
try:
    project = rf.workspace("vismayaks").project("insect-detection-in-grains")
    version = project.version(4)  # Use version 5 of your model
    model = version.model
except Exception as e:
    st.error(f"Error loading project or model: {e}")
    st.stop()

# Streamlit app interface
st.title("Rice Grain Pest Detection")

# File uploader for images
uploaded_file = st.file_uploader("Upload an Image", type=["jpg", "jpeg", "png"])

if uploaded_file:
    # Open the image
    image = Image.open(uploaded_file)
    image_array = np.array(image)

    st.subheader("Original Image")
    st.image(image, caption="Uploaded Image", use_container_width=True)

    # Save the uploaded image to a temporary file
    with tempfile.NamedTemporaryFile(delete=False, suffix=".jpg") as tmpfile:
        tmpfile.write(uploaded_file.getvalue())
        temp_image_path = tmpfile.name

    # Get detections using Roboflow API
    try:
        prediction = model.predict(temp_image_path).json()
    except Exception as e:
        st.error(f"Error during prediction: {e}")
        st.stop()

    # Display prediction results
    st.subheader("Detected Objects")

    # Process the image with bounding boxes and labels
    for item in prediction['predictions']:
        x = int(item['x'])
        y = int(item['y'])
        width = int(item['width'])
        height = int(item['height'])
        confidence = item['confidence']
        label = item['class']

        # Draw rectangle and label
        top_left = (x - width // 2, y - height // 2)
        bottom_right = (x + width // 2, y + height // 2)
        cv2.rectangle(image_array, top_left, bottom_right, (0, 255, 0), 2)

        label_text = f"{label} ({confidence:.2f})"
        cv2.putText(image_array, label_text, (top_left[0], top_left[1] - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Convert the image to RGB for Streamlit
    image_with_detections = cv2.cvtColor(image_array, cv2.COLOR_BGR2RGB)

    # Display image with detections
    st.image(image_with_detections, caption="Image with Detections", use_container_width=True)
```
